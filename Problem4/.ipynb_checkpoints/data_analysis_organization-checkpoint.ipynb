```python
#there are 120 breeds to classify into
#let us organize the data on the basis of labels
import os
os.mkdir('dataset/organized')
```

```python
os.mkdir('dataset/organized/training')
os.mkdir('dataset/organized/validating')
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size=0.2)
```

```python
for label in labels:
    parent1 = 'dataset/organized/training/'
    parent2 = 'dataset/organized/validating/'
    label_dir = label
    path1 = os.path.join(parent1, label_dir)
    path2 = os.path.join(parent2, label_dir)
    os.mkdir(path1)
    os.mkdir(path2)
```

```python
import shutil
img_test = test.iloc[:, 0].values
label_test = test.iloc[:, 1].values
```

```python
for i in range(0, len(img_test)):
    cur_path = 'dataset/train/'+img_test[i]+'.jpg'
    path = 'dataset/organized/validating/'+label_test[i]+'/'+img_test[i]+'.jpg'
    shutil.copyfile(cur_path, path)
```

```python
img_train = train.iloc[:,0].values
label_train = train.iloc[:,1].values
```

```python
for i in range(0, len(img_train)):
    cur_path = 'dataset/train/'+img_train[i]+'.jpg'
    path = 'dataset/organized/training/'+label_train[i]+'/'+img_train[i]+'.jpg'
    shutil.copyfile(cur_path, path)
```

```python
import tensorflow as tf
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(
                rescale=1./255,
                shear_range=0.2,
                zoom_range=0.2,
                horizontal_flip=True)
```

```python
training_set = train_datagen.flow_from_directory(
                'dataset/organized/training',
                target_size=(128,128),
                batch_size=32,
                class_mode='categorical')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
test_set = test_datagen.flow_from_directory(
            'dataset/organized/validating',
            target_size=(128,128),
            batch_size=32,
            class_mode='categorical')
```

```python
cnn = tf.keras.models.Sequential()
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[128,128,3]))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=256, activation='relu'))
cnn.add(tf.keras.layers.Dense(units=256, activation='relu'))
cnn.add(tf.keras.layers.Dense(units=256, activation='relu'))
cnn.add(tf.keras.layers.Dense(units=120, activation='softmax'))
```

```python
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn.fit(x=training_set, validation_data=test_set, epochs=100)
```

```python
import numpy as np
from keras.preprocessing import image
test_image1 = tf.keras.utils.load_img('dataset/test/0a0b97441050bba8e733506de4655ea1.jpg', target_size=(64,64))
test_image1 = tf.keras.utils.img_to_array(test_image1)
test_image1 = np.expand_dims(test_image1, axis=0)
training_set.class_indices
```

```python
result = cnn.predict(test_image1)
result = result[0].tolist()
sum(result)
```

```python
sample = pd.read_csv('dataset/sample_submission.csv')
sample
```

```python
submission = pd.DataFrame(columns=[sample.columns])
```

```python
ids = sample['id'].values
```

```python
for i in range(0, len(ids)):
    row = [ids[i]]
    test_image = tf.keras.utils.load_img('dataset/test/' + ids[i] + '.jpg', target_size=(64,64))
    test_image = tf.keras.utils.img_to_array(test_image)
    test_image = np.expand_dims(test_image, axis=0)
    result = cnn.predict(test_image)
    result = result[0].tolist()
    row = row + result
    submission.loc[i] = row
```
